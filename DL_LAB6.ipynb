```python
# ==============================
# LAB 6: ATTENTION (IMPROVED)
# ==============================

import torch
import torch.nn.functional as F
import seaborn as sns
import matplotlib.pyplot as plt

sentence = ["Stock", "prices", "are", "increasing", "rapidly"]

# Simulated attention (realistic pattern)
weights = torch.tensor([0.05, 0.2, 0.1, 0.4, 0.25])
weights = F.softmax(weights, dim=0).numpy()

sns.heatmap([weights], annot=True, xticklabels=sentence, cmap="Blues")

plt.title("Attention Heatmap")
plt.show()
```
